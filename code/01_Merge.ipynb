## 데이터 세트 별 기준 컬럼 unique value 확인

```python
merge_on = ['년도', '분기', '상권유형', '상권', '업종']

df_name = ['추정매출', '점포현황', '소득소비', '상주인구', '생활인구', '직장인구', '아파트', '상권변화지표']
df_list = [target, store, incomeConsumption, pop0, pop1, pop2, apartment, vari]
```

```python
for col in merge_on :
    print(col)
    for i, df in enumerate(df_list) :
        try : print(f"{df_name[i]} : {df[col].nunique()}")
        except : print(f"{df_name[i]} : 칼럼 없음")
    print("\n")
```

## merge

```python
df_list = [target, store, incomeConsumption, pop0, pop1, pop2, apartment, vari]
merge_on = ['년도', '분기', '상권유형', '상권', '업종']

for i in range(len(df_list)) :
    if i == 0 :
        continue
    elif i == 1 :
        df_merge = pd.merge(df_list[i-1], df_list[i], how = 'inner', on = merge_on)
    else :
        df_merge = pd.merge(df_merge, df_list[i], how = 'inner', on = merge_on[:-1])
```

```python
for i, j in zip(df_merge.columns, df_merge.isnull().mean()) :
    print(i, ":", j)
```
